## Column Name Cleaning

```python
# Standardizing column names
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})

# Verify changes
df.columns
```

## Feature Engineering

```python
# Creating age groups
df['age_group'] = pd.cut(
    df['age'],
    bins=[0,19,35,55,100],
    labels=['Teen','Young','Middle Age','Senior']
)

df[['age','age_group']].head()
```

## Handling Missing Values

```python
# Checking missing values
df.isnull().sum()[df.isnull().sum() > 0]
```

```python
# Fill missing review ratings using category median
category_median = df.groupby('category')['review_rating'].median()
df['review_rating'] = df['review_rating'].fillna(df['category'].map(category_median))
```

```python
# Checking for remaining missing values after imputation
df.isnull().sum()[df.isnull().sum() > 0]
```

## Removing Redundant Columns

```python
# Comparing both columns
df[["discount_applied", "promo_code_used"]].head(5)
```

```python
# Check if promo_code_used and discount_applied are identical
(df['promo_code_used'] == df['discount_applied']).all()
```

```python
# Drop redundant column
df = df.drop(columns=['promo_code_used','frequency_of_purchases'])

# Verify
df.columns
```
